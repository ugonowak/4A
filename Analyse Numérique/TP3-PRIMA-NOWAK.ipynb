```python
import numpy as np
```

```python
def puissance(A, q0):
    x = np.zeros(len(A))
    lam = np.zeros(len(A))
    q0inf = np.zeros(len(A))
    q0_2 = np.zeros(len(A))
    cpt = 0
    while (cpt < 1000):
        for i in range(len(A)):
            x = np.dot(A, q0)
            if abs(q0[i]) > 1e-10:     
                lam[i] = x[i] / q0[i]
            q0[i] = (1/np.max(x)) * x[i]
            q0_2[i] = (1/np.linalg.norm(x)) * x[i]
        cpt += 1
    return q0, q0_2, lam    
```

```python
def spectre(A):
    valeurs_propres = np.linalg.eig(A)
    return valeurs_propres[0]
```

```python
def puissance_shift(A, q0, mu):
    x = np.zeros(len(A))
    lam = np.zeros(len(A))
    cpt = 0
    I = np.eye(len(A))
    B = A - mu * I
    
    while (cpt < 1000):
        x = np.zeros(len(A))
        x = np.linalg.solve(B, q0)
        gamma = x[np.argmax(np.abs(x))]
        for i in range(len(A)):
            q0[i] = x[i] / gamma
        cpt += 1
        
    return q0, lam
```

Avec cette méthode cela fonctionne à la différence de la méthode de la puissance classique
